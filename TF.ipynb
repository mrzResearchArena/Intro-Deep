#### The Equation to Solve:
###### $$F(w) = w^2 - 10w + 25$$
###### $$F'(w) = 2w - 10$$
###### As, we know that F'(w)=0, then w=5.

#### Varable Initialization:

```python
w = tf.Variable(0, dtype=tf.float32)
# cost = tf.add(tf.add(w**2, tf.multiply(-10.0,w)), 25)
cost = (w**2.0) - (10*w) + 25
```

#### Gradient Descnt Optimization (GD, Momentum, RMSprop, Adam):

```python
train = tf.train.GradientDescentOptimizer(0.01).minimize(cost)
# train = tf.train.MomentumOptimizer(learning_rate=0.01, momentum=0.9).minimize(cost)
# train = tf.train.RMSPropOptimizer(learning_rate=0.01).minimize(cost)
# train = tf.train.AdamOptimizer(learning_rate=0.1).minimize(cost)
```

```python
init = tf.global_variables_initializer()
session = tf.Session()
session.run(init)
print(session.run(w))
```

```python
session.run(train)
print(session.run(w))
```

```python
C = 1
for i in range(500):
    session.run(train)
    print('{}: {}'.format(C, session.run(w)))
    C += 1 
```
